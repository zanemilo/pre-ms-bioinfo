### Step One<a id="step_one"></a>:

The first step is to be in the working directory the venv will be located. So ensure it is somewhere easy to access, clearly labeled. etc.

Given the venv is not set up, follow the steps below:

1. open command line to the directory the venv will be in
2. run the following python code:
       ```python venv -m venv``` or ```python3 venv -m venv```
3. activate the venv with the following:
        MacOS: ```source venv/bin/activate```
        Windows: ```.venv\Scripts\activate.bat```


### Step Two<a id="step_two"></a>:

The next step here is to install R through this **[Link](https://posit.co/download/rstudio-desktop/)**

Confirm R is installed by typing ```R``` in the command line.


### Step Three<a id="step_three"></a>:

At this point we need to download the R kernel in the **venv** in order to run R in jupyter notebook.
To do so, within the **venv**, type **```R```**, then we will need to type the follwing lines into the command line:

1. ```install.packages("devtools")```
2. ```install.packages("IRkernel")```
3. ```devtools::install_github("IRkernel/IRkernel")```
4. ```IRkernel::installspec()```
5. ```q()```-> It is safe to select 'N' to not save on quit.

src -> [Link](https://stackoverflow.com/questions/57870575/install-and-run-r-kernel-for-jupyter-notebook)


### Step Four<a id="step_four"></a>:

Verify the kernel was installed in the venv, type the following in to command line:
```jupyter kernelspec list```

The below should be output if Python and R kernels are installed in the venv for Jupyter:

<code>Available kernels:
    python3    /path/to/python3/share/jupyter/kernels/python3
    ir         /path/to/r/share/jupyter/kernels/ir<code/>
