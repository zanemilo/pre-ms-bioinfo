```python
import matplotlib.pyplot as plt
```

First, ensure to import matplotlib. Pyplot is a sub-module within Matplotlib that simplifies creating visualization in Python with MATLAB-like syntax.

### Line Plot

```python
x = [12, 24, 35, 99, 103]
y = [24, 35, 46, 77, 83]

plt.plot(x,y)
plt.show()
```

plt.plot(x,y) creates the line plot with x as x-axis & y as y-axis.

plot.show() is the method to display the visualization.

### Pie Chart

```python
labels = ['Chocolate','Vanilla', 'Strawberry', 'Other']
survey_results = [47, 23, 33, 15]

plt.pie(survey_results, labels=labels, autopct='%1.1f%%')
plt.show()
```

plt.pie(params) -> creates each pie slice size based on survey_results values, labels from labels corresponding to each result and uses autopct='%1.1f%%' to instruct the value to round and display with one decimal place.

### Bar Chart

```python
categories = ['Lion', 'Cheetah', 'Cat', 'Leopard']
vals = [x + 2 * x + 1 for x in range(4)]  # Quick list comprehension to get vals populated for example

plt.bar(categories, vals)
plt.show()
```

plt.bar(params) draws a vertical bar graph using given categories and values

### Scatter Plot

```python
x = [x for x in range(5)]
y = [y**3 for y in x]

plt.scatter(x,y)
plt.show()
```

plt.scatter(params) plots data points as dots at the param coordinates.

### Histogram

```python
import numpy as np
```

Ensure to import numpy to generate random values for the histogram.

```python
data = np.random.randn(1000)
plt.hist(data, bins=50)
plt.show()
```

np.random.randn(x) -> generates x random numbers from a standard normal distribution.
plt.hist(params) - > creawtes a histogram of the data with intervals based on bin param.

### Stack Plot

```python
days = [1, 2, 3, 4, 5]
studying = [7, 8, 6, 11, 7]
playing = [8, 5, 7, 8, 13]

plt.stackplot(days, studying, playing, colors =['r', 'c'])


plt.xlabel('Days') # Labels x-axis

plt.ylabel('No of Hours') # Labels y-axis

plt.title('Representation of Study and Playing w/ resp. to Days') # Label Plot



plt.show()
```

Stack plots stack data values ontop one another to show cumulative data. In this case we see Days along the x-axis and Hours along the y-axis. The data stacked is the time spent studying and playing on each day.

A good example -> chess.com uses the stack plot to show the number of each quality of move one made on a given day.

### Stem Plot

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))


plt.stem(x, y)
plt.show()
```

Stem plots are useful visualizating discrete data along a number line. AKA Stem and Leaf plot where Leaf is the individual data point, the stem is the line extending from the baseline (red).

### Box Plot

```python
np.random.seed(11)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize = (10, 7))

plt.boxplot(data)

plt.show()
                 
```

### Error Plot

```python
x = [x*np.random.normal(10) for x in range(10)]
y = [y*np.random.normal(5) for y in range(10)]

y_error = 2.2
x_error = 6.3

plt.plot(x,y)

plt.errorbar(x,y,yerr = y_error, xerr = x_error, fmt='o')
```

Error plots display variability or uncertainty for each data point in the dataset. Visualizes measurement errors, confidence intervals, standard deviations, etc.

### Violin Plot

```python
uniform_dist_dataset = np.arange(-100, 100)

normal_dist_dataset = np.random.normal(size=100)*50

fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize =(9,4), sharey = True)

ax1.set_title("Uniform Distribution")
ax1.set_ylabel("Observed Values")
ax1.violinplot(uniform_dist_dataset)

ax2.set_title("Normal Distribution")
ax2.violinplot(normal_dist_dataset)
```

Subplots are smaller axes (plots) arranged within a single figure allowing users to compare data side by side or stack visualizations for better insights. plt.subplots() method provides an easy way to define the layout of subplots.

### 3D Plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]

fig = plt.figure()

ax = plt.axes(projection = '3d')
ax.plot3D(z, y, x)
plt.show()
```
