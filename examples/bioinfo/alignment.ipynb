## **Files**

### FASTQ:
The Raw reads data 
- Identifier
- Sequence
- Seperator
- QC score encoded in ASCII
- May be trimmed post QC check -> new trimmed file. Do not modify base file ever.


### FASTQC:
Output from QC check
- Quality per base
- Adapter contamination
  

### FASTA:
- Chromosome sequences
- Used for alignment (by STAR, for example)


### GTF:
- Gene locations
- Exons
- Transcripts
- Used by: STAR (splice awareness) & Downstream counting


### BAM File:
- Each readâ€™s genomic position
- Mapping quality
- CIGAR strings (splicing info)


### BAM Index:
- Fast random access
- Required for visualization


### Star Log Files
- % mapped reads
- Unique vs multi-mapped
- Splice junction stats


### Count matrices:
- csv
- tsv
